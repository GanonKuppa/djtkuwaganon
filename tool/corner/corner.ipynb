```python
%matplotlib inline
import numpy as np
from scipy.optimize import curve_fit
from matplotlib import pyplot as plt
import math

TIMESTAMP_COL = 0
WALL_L_COL = 28
WALL_R_COL = 29
THREE_SECTION_AHEAD_DIST_COL = 52

wall_corner_threshold_on_l = 200
wall_corner_threshold_off_l = 160
wall_corner_threshold_on_r =  200
wall_corner_threshold_off_r = 160

diag_corner_threshold_on_r = 400
diag_corner_threshold_off_r = 300

diag_corner_threshold_on_l = 400
diag_corner_threshold_off_l = 300
```

# 壁無し直進時壁切れ調整

```python
# ログ用csvファイルの読み込み(壁無し)

path = "C:/Users/qtfdl/Desktop/log_kushi_r_2.csv"
csv_data = np.loadtxt(path, delimiter=",", dtype = float, skiprows=1)
timestamp = csv_data[:, TIMESTAMP_COL]
wall_r = csv_data[:, WALL_L_COL]
wall_l = csv_data[:, WALL_R_COL]
three_section_ahead_dist = csv_data[:, THREE_SECTION_AHEAD_DIST_COL]

```

```python
# 左壁切れ(壁無し)

plt.figure(figsize=(20,10))
plt.xticks(np.arange(0.045, 0.28, step=0.01))
plt.yticks(np.arange(0, 4000, step=20))
plt.ylim(0, 1000)
plt.grid()

plt.vlines(x=0.09, ymin=0, ymax=4000, color='r', linestyles='dotted')
plt.vlines(x=0.135, ymin=0, ymax=4000, color='b', linestyles='dotted')
plt.vlines(x=0.18, ymin=0, ymax=4000, color='r', linestyles='dotted', lw=5)
plt.vlines(x=0.27, ymin=0, ymax=4000, color='r', linestyles='dotted')

plt.hlines(y=wall_corner_threshold_on_l, xmin=0, xmax=0.27, color='g', linestyles='solid')
plt.hlines(y=wall_corner_threshold_off_l, xmin=0, xmax=0.27, color='g', linestyles='solid')

plt.plot(0.09 * 3 - three_section_ahead_dist, wall_l, lw=1, label="wall_l")
plt.legend()
plt.show()
```

```python
# 右壁切れ壁無し

plt.figure(figsize=(20,10))
plt.xticks(np.arange(0.045, 0.28, step=0.01))
plt.yticks(np.arange(0, 4000, step=20))
plt.ylim(0, 1000)
plt.grid()

plt.vlines(x=0.09, ymin=0, ymax=4000, color='r', linestyles='dotted')
plt.vlines(x=0.135, ymin=0, ymax=4000, color='b', linestyles='dotted')
plt.vlines(x=0.18, ymin=0, ymax=4000, color='r', linestyles='dotted', lw=5)
plt.vlines(x=0.27, ymin=0, ymax=4000, color='r', linestyles='dotted')

plt.hlines(y=wall_corner_threshold_on_r, xmin=0, xmax=0.27, color='g', linestyles='solid')
plt.hlines(y=wall_corner_threshold_off_r, xmin=0, xmax=0.27, color='g', linestyles='solid')


plt.plot(0.09 * 3 - three_section_ahead_dist, wall_r, lw=1, label="wall_r")
plt.legend()
plt.show()
```
